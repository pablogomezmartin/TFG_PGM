```python
import matplotlib.pyplot as plt
import pandas as pd

# Leer el archivo y procesar los datos
with open('prueba.txt', 'r') as file:
    lines = file.readlines()

# Crear un diccionario para almacenar las combinaciones y sus frecuencias
combinations = {}

for line in lines:
    # Ignorar lneas vacas
    if line.strip() == '':
        continue

    # Obtener la combinacin de variables (ignorar el ltimo valor que es 0 o 1)
    combination = ','.join(line.split(',')[:-1])

    # Incrementar la frecuencia de la combinacin en el diccionario
    if combination in combinations:
        combinations[combination] += 1
    else:
        combinations[combination] = 1

# Crear un DataFrame a partir del diccionario
df = pd.DataFrame(list(combinations.items()), columns=['Combination', 'Frequency'])

# Ordenar el DataFrame por frecuencia
df = df.sort_values('Frequency', ascending=False)

# Seleccionar las primeras 7 combinaciones
top_7 = df.iloc[:7].copy()

# Sumar las frecuencias del resto y aadir una fila 'Otros'
others = pd.DataFrame([['Otros', df.iloc[7:]['Frequency'].sum()]], columns=['Combination', 'Frequency'])
top_7 = pd.concat([top_7, others])

# Crear el grfico de pastel
fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(top_7['Frequency'], labels = top_7['Combination'], autopct='%1.1f%%', textprops=dict(color="w"))
ax.legend(wedges, top_7['Combination'], title="Combinaciones", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=8, weight="bold")
ax.set_title('Frecuencia de combinaciones de variables')
plt.show()
```
